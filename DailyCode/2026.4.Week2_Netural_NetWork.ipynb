```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X=np.array([
    [100,5],
    [110,13],
    [120,19],
    [140,29]
])
Y=np.array([1, 0, 0, 1])
```

```python
model=Sequential([
    Dense(units=25,activation='relu'),
    Dense(units=25,activation='relu'),
    Dense(units=10,activation='softmax')
])
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
model.compile(optimizer='adam',loss=SparseCategoricalCrossentropy(from_logits=True))
model.fit(X,Y,epochs=100)
```

```python
logits=model(X)
f_x=tf.nn.softmax(logits)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1. еҠ иҪҪ MNIST ж•°жҚ®йӣҶ (зі»з»ҹдјҡиҮӘеҠЁдёӢиҪҪ)
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. еҪ’дёҖеҢ– (жҠҠеғҸзҙ д»Һ 0-255 еҸҳжҲҗ 0-1пјҢиҝҷдёҖжӯҘйқһеёёйҮҚиҰҒпјҒ)
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. жҗӯе»әжЁЎеһӢ (иҝҷе°ұжҳҜдҪ д№ӢеүҚз”»зҡ„йӮЈз§Қз»“жһ„)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)), # жҠҠ 28x28 зҡ„зҹ©йҳөжӢүе№іжҲҗ 784 з»ҙеҗ‘йҮҸ
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),                  # йҳІжӯўиҝҮжӢҹеҗҲзҡ„е°ҸжҠҖе·§
    layers.Dense(10, activation='softmax') # 10дёӘж•°еӯ—пјҢз”Ё softmax
])

# 4. зј–иҜ‘е’Ңи®ӯз»ғ
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("жӯЈеңЁи®ӯз»ғпјҢиҜ·зЁҚеҖҷ...")
model.fit(x_train, y_train, epochs=5) # и·‘ 5 иҪ®е°ұеӨҹ 97% еҮҶзЎ®зҺҮдәҶ

# 5. йҡҸдҫҝжҢ‘дёҖеј жөӢиҜ•еӣҫзңӢзңӢз»“жһң
plt.imshow(x_test[0], cmap='gray')
plt.show()

prediction = model.predict(x_test[0:1])
print(f"жЁЎеһӢиҜҙиҝҷдёӘж•°еӯ—жҳҜ: {np.argmax(prediction)}")
```
