### **Now I will try to Accomplish the Gradient Decent**

### **Module is linear Regression y=wx+b, loss function is (y-yi)**

```python
import numpy as np
x=np.array([1.0,2.0,3.0,4.0])
y=np.array([300.0,500.0,600.0,800.0])
```

```python
import numpy as np
#计算损失函数，cost=f_wb - y
def compute_cost(x,y,w,b):
    m=x.shape[0]
    f_wb=w*x + b
    cost=(1/(2*m))*np.sum((f_wb-y)**2)
    return cost
```

```python
def gradient_decent(x,y,w,b):
    m=x.shape[0]
    f_wb=w*x + b
    dj_dw=1/m*np.sum((f_wb-y)*x)#np自动会把矢量对应的东西加起来
    dj_db=1/m*np.sum(f_wb-y)
    return dj_dw,dj_db
```

```python
w,b=0.0,0.0
a=0.01
iteration=1000
for i in range(iteration):
    dw,db=gradient_decent(x,y,w,b)
    w=w-a*dw
    b=b-a*db
    if i % 100 == 0:
        print(f"Iteration {i}: Cost = {compute_cost(x, y, w, b)}")

print(f"最终预测模型: y = {w:.2f}x + {b:.2f}")
```
