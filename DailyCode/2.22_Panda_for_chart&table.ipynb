### 4.1 Adding columns

#### 4.1.1 **Assign()**

```python
import numpy as np
import pandas as pd
```

```python
score1 =np.random.randint(1,1000,size=5)
score2 =np.random.randint(1,1000,size=5)
```

```python
series1 =pd.Series(score1,index=np.arange(0,5))
series2 =pd.Series(score1,index=np.arange(0,5))
series1
```

```python
df_eg= df_eg.assign(score1 = series1,score2=series2)
```

```python
df_eg #记得赋值一下，不然不会修改原表格
```
