#### 1.1 Read serveral csv

```python
# Link structure : Root + season + league
'https://www.football-data.co.uk/mmz4281/'+'2526'+'/'
```

```python
# Create a root
root ='https://www.football-data.co.uk/mmz4281/'
```

```python
import pandas as pd
import requests
import io

# 1. 定义根目录
root = 'https://www.football-data.co.uk/mmz4281/'
leagues = ['E0', 'E1', 'E2','E3','E4','EC']
frame = []

# 2. 【关键】伪装成普通的 Chrome 浏览器
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

for league in leagues:
    url = f"{root}2122/{league}.csv"  # 咱们先用 2122 赛季测试
    print(f"正在伪装抓取: {url}")

    try:
        # 使用 requests 发送带“迷彩服”的请求
        response = requests.get(url, headers=headers)

        # 检查是否请求成功
        if response.status_code == 200:
            # 使用 io.StringIO 把抓到的文本变成 Pandas 能读的“假文件”
            # 记得用你之前问的那个 unicode_escape
            df = pd.read_csv(io.StringIO(response.text), encoding='utf-8-sig')
            frame.append(df)
            print(f"✅ {league} 成功读入！")
        else:
            print(f"❌ 还是被发现是机器人了，错误码: {response.status_code}")

    except Exception as e:
        print(f"💥 发生意外错误: {e}")

# 3. 合并数据
if frame:
    final_df = pd.concat(frame)
    print(f"🎉 搞定！总共抓到了 {len(final_df)} 行比赛数据。")
```

```python
len(frame)
```

```python
if frame:
    # 2. 针对第一个 DataFrame 的列名进行清洗
    # 我们只清洗第一个，因为合并时 Pandas 以它为准
    frame[0].columns = [col.replace('ï»¿', '') for col in frame[0].columns]

    # 3. 甚至可以更保险，直接强制改第一个 DataFrame 的第一列名字
    frame[0].rename(columns={frame[0].columns[0]: 'Div'}, inplace=True)
```

#### 1.2 **Multiple Seasons**

```python
import pandas as pd
root = 'https://www.football-data.co.uk/mmz4281/'
leagues = ['E0', 'E2', 'E3']
frames = [] # 注意博主这里用的是 frames，复数

for league in leagues:
    for season in range(15, 21): # 爬取 1516 到 2021 赛季
        url = root + str(season)+ str(season+1) + '/' + league + '.csv'
        df = pd.read_csv(url, encoding='unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
```

#### 1.3 **Organize all the data with Dic**

```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish La Liga': 'SP1',
    'Spanish Segunda Division': 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0',
    'English League 1': 'E2',
    'English League 2': 'E3',
}
```

```python
for league in dict_countries:
    print(league)
```

```python
dict_historical_data={}

for league in dict_countries:
    frames=[]
    for season in range(15, 21): # 爬取 1516 到 2021 赛季
        url = root + str(season)+ str(season+1) + '/' + dict_countries[league] + '.csv'
        df = pd.read_csv(url, encoding='unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat=pd.concat(frames)
    dict_historical_data[league]=df_concat
```
