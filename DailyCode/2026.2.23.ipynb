## 4.2 **Math Operation()**

#### 4.2.1 **Operations in columns**

```python
from random import random

import pandas as pd
import numpy as np

# 构造一个模拟咖啡馆的业务数据
data = {
    'Item': ['Latte', 'Americano', 'Cappuccino', 'Mocha', 'Matcha', 'Cold Brew'],
    'Unit_Price': [32.0, 25.0, 30.0, 35.0, 38.0, 28.0],     # 单价 (float)
    'Quantity': [10, 15, 8, 12, 0, 20],                    # 销量 (int)
    'Cost_Per_Cup': [12.5, 8.0, 11.0, 14.0, 15.0, 9.0],    # 每杯成本 (float)
    'Discount_Rate': [0.9, 1.0, 0.85, 1.0, 0.9, 0.95],     # 折扣率 (0.9表示9折)
    'Rating': [4.8, 4.5, 4.2, 4.9, np.nan, 4.7]            # 顾客评分 (带有一个NaN空值)
}

df_cafe = pd.DataFrame(data)
print("--- AI 咖啡馆经营简表 ---")
print(df_cafe)
```

```python
df_cafe['Quantity'].sum()
```

```python
df_cafe['Quantity'].count()
df_cafe['Quantity'].mean()
df_cafe['Quantity'].std()# std 是标准差的意思，衡量波动，和方差一个样
df_cafe['Quantity'].max()
df_cafe['Quantity'].min()
```

```python
df_cafe.describe()
#一个函数抵得上上面的所有
```

#### 4.2.2 **Operations in row**

```python
df_cafe
```

```python
df_cafe['Unit_Price']+df_cafe['Quantity']+df_cafe['Cost_Per_Cup']
```

```python
df_cafe['Raw_Revenue'] = df_cafe['Unit_Price'] * df_cafe['Quantity']
df_cafe.round(2)#这是保留n为小数
```

```python
df_cafe['Rating'].value_counts(normalize=True)#前者返回出现次数，后者若等于True返回百分比
```

#### 4.3 **Sort Operation()**

```python
df_cafe.sort_values(['Rating','Quantity'],ascending=False,inplace=True)
df_cafe
```

```python
df_cafe.sort_values(['Item'],ascending=False,inplace=False,key = lambda col:col.str.lower())
df_cafe
```
