## **Groupby And Aggregates**

```python
import pandas as pd
import numpy as np
import random

# 1. 原始基因库
companies = ['Apple', 'HP', 'Acer', 'Asus', 'Dell', 'Lenovo', 'MSI', 'Toshiba']
types = ['Ultrabook', 'Notebook', 'Gaming', '2 in 1 Convertible', 'Workstation']
cpus = ['Intel i5 2.3GHz', 'Intel i7 2.7GHz', 'Intel i3 6006U', 'AMD A6-9220', 'Intel i7 7500U', 'Intel i5 8250U']
rams = ['4GB', '8GB', '16GB', '32GB']
memories = ['128GB SSD', '256GB SSD', '512GB SSD', '1TB HDD', '500GB HDD', '128GB Flash']
screens = ['1366x768', 'Full HD 1920x1080', 'IPS Panel Retina 2560x1600', '4K Ultra HD']
os_list = ['Windows 10', 'macOS', 'Linux', 'No OS', 'Android']
status_list = ['In Stock', 'Out of Stock', 'Pre-order']

rows = []
for i in range(1, 1001):
    company = random.choice(companies)
    # 逻辑微调：如果是 Apple 强制 macOS
    os = 'macOS' if company == 'Apple' else random.choice(['Windows 10', 'Linux', 'No OS'])

    rows.append({
        'laptop_ID': i,
        'Company': company,
        'Product': f"{company} Model-{random.randint(100, 999)}",
        'TypeName': random.choice(types),
        'Inches': random.choice([13.3, 14.0, 15.4, 15.6, 17.3]),
        'ScreenResolution': random.choice(screens),
        'Cpu': random.choice(cpus),
        'Ram': random.choice(rams),
        'Memory': random.choice(memories),
        'Price_Euros': round(random.uniform(300, 3000), 2),      # 1. 价格 (小数)
        'OpSys': os,                                             # 2. 操作系统
        'Weight_kg': round(random.uniform(0.9, 4.5), 2),        # 3. 重量 (小数)
        'Stock_Status': random.choice(status_list),             # 4. 库存状态 (分类)
        'Rating': random.randint(1, 5)                          # 5. 用户评分 (1-5星)
    })

df_laptops = pd.DataFrame(rows)
df_laptops.head(7)
```

```python
df_laptops=df_laptops[['Company','Product','TypeName','ScreenResolution','Ram','Price_Euros','Rating']]
df_laptops.nunique()
```

### **1: Agg()**

```python
df_laptops[['Price_Euros','Rating']].agg('sum')
df_laptops.agg('sum')# 也有mean，但是这里有str所以用不了
```

```python
df_laptops.agg({'Rating':['mean','sum'],
                'Price_Euros':['max','sum']})
```

```python
df_laptops[['Rating','Price_Euros']].agg('sum',axis=1)
```

```python
high_R=df_laptops['Rating']>=3
low_R=df_laptops['Rating']<3
df_laptops[low_R].count()
```

### **Group_by method：可以进行好玩的分组**

```python
g1=df_laptops.groupby('Company')
g1
```

```python
g1.groups.keys()  #给你分好组了
```

```python
g1.get_group('Apple')
```

```python
df_laptops.groupby('Price_Euros').min()
```

```python
df_laptops.groupby('Rating')['Price_Euros'].mean()
```

```python
df_laptops.groupby('Price_Euros')['Rating'].agg('sum')#括号里还能加其他的，比如max和min，mean，count等等，只不过这个表格不能用
```
